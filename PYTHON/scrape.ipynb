```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
site = requests.get('https://coinmarketcap.com/')
soup = BeautifulSoup(site.content,'html.parser')

names = []
symbols = []
price = []

data = soup.find_all('p', class_='sc-1eb5slv-0 iJjGCS')
for i in data:
    names.append(i.get_text())
    
data = soup.find_all('p', class_='sc-1eb5slv-0 gGIpIK coin-item-symbol')
for i in data:
    symbols.append(i.get_text())

sym = soup.tbody
sym = sym.find_all(class_='cmc-link')
for i in sym:
    a = i.find_all('span')
    if a != []:
        names.append(a[1].get_text())
        symbols.append(a[2].get_text())
```

```python
price = []
lowvaluein24h = []
highvaluein24h = []
circulatingsupply= []
volume = []
marketcap = []
dilutedmarketcap = []
for coin in names:
    try:
        name = coin
        name = name.replace(' ','-')
        site = requests.get(f'https://coinmarketcap.com/currencies/{name.lower()}/')
        soup = BeautifulSoup(site.content,'html.parser')
        a=soup.find_all('div',class_='priceValue___11gHJ')
        price.append(a[0].get_text())

        a=soup.find_all('span',class_='highLowValue___GfyK7')
        lowvaluein24h.append(a[0].get_text())
        highvaluein24h.append(a[1].get_text())

        a=soup.find_all('div',class_='statsValue___2iaoZ')

        marketcap.append(a[0].get_text())
        dilutedmarketcap.append(a[1].get_text())
        volume.append(a[2].get_text())
        circulatingsupply.append(a[4].get_text())        
        
        
    except Exception as e:
        price.append('')
        lowvaluein24h.append('')
        highvaluein24h.append('')
        circulatingsupply.append('')
        volume.append('')
        marketcap.append('')
        dilutedmarketcap.append("Market data is untracked.This project is featured as an 'Untracked Listing'")
df = pd.DataFrame(columns = ['Name', 'Symbol', 'Price', 'Marketcap', 'Volume', 'Diluted marketcap', 'Circulating Supply', 'Low value (24h)', 'High value (24h)' ])

df['Name'] = names
df['Symbol'] = symbols
df['Price'] = price
df['Marketcap'] = marketcap
df['Volume'] = volume
df['Diluted marketcap'] = dilutedmarketcap
df['Circulating Supply'] = circulatingsupply
df['Low value (24h)'] = lowvaluein24h
df['High value (24h)'] = highvaluein24h
df.index += 1
df.to_csv('cryptocurrency.csv')
df
```
