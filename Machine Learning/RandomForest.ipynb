```python
# Data train menggunakan data dari tugas decision tree yang berjumlah 12 data
# Data test menggunakan data dari tugas ensemble yang berjumlah 7 data
# hasil akurasi 100%

import pandas as pd
import numpy as np

age = '25-30,25-30,31-40,>40,>40,31-40,31-40,25-30,>40,>40,25-30,25-30'.split(',')
income = 'High,High,High,Medium,Low,Low,Medium,Low,Medium,High,Medium,Medium'.split(',')
credit = 'Fair,Excellent,Fair,Fair,Fair,Excellent,Excellent,Fair,Excellent,Excellent,Excellent,Fair'.split(',')
buys = 'No,No,No,Yes,Yes,Yes,Yes,Yes,No,Yes,Yes,No'.split(',')

ageTest = '25-30,25-30,>40,>40,31-40,31-40,>40'.split(',')
incomeTest = 'High,High,Medium,Low,Low,Medium,High'.split(',')
creditTest = 'Fair,Excellent,Fair,Fair,Excellent,Excellent,Excellent'.split(',')
buysTest = 'No,No,Yes,Yes,Yes,Yes,Yes'.split(',')

dataset ={'Age':age,'Income':income,'Credit_Rating':credit,'buys_car':buys}
dataTest = {'Age':ageTest,'Income':incomeTest,'Credit_Rating':creditTest,'buys_car':buysTest}
dTrain = pd.DataFrame(dataset,columns=['Age','Income','Credit_Rating','buys_car'])
dTest = pd.DataFrame(dataTest,columns=['Age','Income','Credit_Rating','buys_car'])

print("Data Train: ")
print(dTrain)
print("\nData Test: ")
print(dTest)

buys_car_map = {'Yes': 1, 'No': 0}
dTrain['buys_car'] = dTrain['buys_car'].map(buys_car_map)
dTest['buys_car'] = dTest['buys_car'].map(buys_car_map)

y_train = dTrain.pop('buys_car')
x_train = dTrain
y_test = dTest.pop('buys_car')
x_test = dTest

features_to_encode = x_train.columns[x_train.dtypes==object].tolist()

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
col_trans = make_column_transformer(
                        (OneHotEncoder(),features_to_encode),
                        remainder = "passthrough"
                        )

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline
clf = RandomForestClassifier(n_estimators=5, random_state=50)
pipe = make_pipeline(col_trans, clf)
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred))
print('RF accuracy: %s' % pipe.score(x_test, y_test))
```
